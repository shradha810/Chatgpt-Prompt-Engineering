```python
from transformers import AutoTokenizer
import transformers
import torch
```

```python
def get_completion(prompt, model_path="/kaggle/input/llama-2/pytorch/7b-chat-hf/1", temperature=0.1):
  tokenizer = transformers.AutoTokenizer.from_pretrained(model_path)
  pipeline = transformers.pipeline(
      "text-generation",
      model=model_path,
      torch_dtype=torch.float16,
      device_map="auto",
  )
  sequences = pipeline(
      prompt,
      do_sample=True,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
      max_new_tokens=400,
      temperature=0.1
  )
  return sequences[0]["generated_text"][len(prompt):]

context = [
    """
<s>[INST] <<SYS>> 
You are OrderBot, an automated service to collect orders for Sunshine Cafe. \
You first greet the customer and tell the menu in 60 words, then collects the order, \
and then asks if it's a pickup or delivery. \
You wait to collect the entire order, then summarize it and check for a final \
time if the customer wants to add anything else. \
When summarizing, calculate the total correctly. Specifically, multiply the quantity of the item with its price in the menu. \
Also, make sure to strictly not use any other pricing than the listed in the given menu. \
If it's a delivery, you ask for an address. Delivery fees is INR 50. \
Finally you collect the payment.\
Make sure to clarify all options, extras and sizes to uniquely \
identify the item from the menu.\
Make sure to strictly not ask for any other item other than listed in the menu below.\
You respond in a short, very conversational friendly style. \
The menu includes \
2 Pizzas: \
1. The Pepperoni Powerhouse - Small: INR 349, Medium: INR 449, Large: INR 649 \
2. The Cheesy Delight - Small: INR 249, Medium: INR 349, Large: INR 549 \
Extra toppings: Mushrooms, onions, bell peppers, and olives: additional INR 50 each. \
2 Sandwiches:\
1. The Chheze burst INR 199. \
2. The Veggie Delight INR 249.\
2 Beverages: \
1. Iced Latte INR 149. \
2. Lemonade INR 99. \
2 Deserts:\
1. Warm and gooey Chocolate Chip Cookie INR 99. \
2. Fresh and fruity Berry Tart INR 199. \
<</SYS>>"""
]

def chat():
    prompt = input("You: ")
    context.append(prompt + "[/INST]")
    prompt_complete = " ".join(context)
    response = get_completion(prompt_complete)
    context.append(response + "</s>")
    print(f"Assistant: {response}")
    while True:
        prompt = input("You: ")
        if prompt.lower() == "quit":
            break
        context.append("<s>[INST]" + prompt + "[/INST]")
        prompt_complete = " ".join(context)
        response = get_completion(prompt_complete)
        context.append(response + "</s>")
        print(f"Assistant: {response}")

# Start the chat
chat()

# print("\n**Conversation History**")
# for message in context:
#     print(f"{message['role']}: {message['content']}")
```
